```python
import pandas as pd 
import matplotlib.pyplot as plt 
import numpy as np
```

```python
gtemp = pd.read_csv('global.csv') 
ctemp = pd.read_csv('cities.csv')
```

```python
glb_mv_avg = gtemp['avg_temp'].rolling(10).mean()
local_mv_avg = ctemp['avg_temp'].rolling(10).mean()
print(glb_mv_avg.min())
print(glb_mv_avg.max())
print(local_mv_avg.min())
print(local_mv_avg.max())
```

```python
plt.plot(gtemp['year'],glb_mv_avg,label='Global')
plt.plot(ctemp['year'],local_mv_avg,label='Kanpur')
plt.legend()
plt.xlabel("Years")
plt.ylabel("Temperature (Â°C)") 
plt.title("Kanpur City Average Temperature")
plt.show()
```
